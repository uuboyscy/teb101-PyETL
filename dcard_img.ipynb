```python
import requests
from urllib import request

headers = {'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_15_3) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/80.0.3987.163 Safari/537.36'}
url = 'https://www.dcard.tw/service/api/v2/forums/photography/posts?limit=30&before=233539423'

res = requests.get(url, headers=headers)

import json

json_str = res.text
json_data = json.loads(json_str)
```

```python
import ssl
ssl._create_default_https_context = ssl._create_unverified_context

for article in json_data:
    article_title = article['title']
    article_url = 'https://www.dcard.tw/f/photography/p/' + str(article['id'])
    article_imgs = article['media'] # List
    article_imgs_list = [i['url'] for i in article_imgs]
    
    print(article_title)
    print(article_url)
    print(article_imgs_list)
    
    for n, img in enumerate(article_imgs_list):
        img_name = article_title + str(n) + '.' + img.split('.')[-1] 
        print(img_name)
        try:
            request.urlretrieve(img, './dcard/{}'.format(img_name))
        except:
            pass
        
    print('==============')
```
