```python
import jieba
import os

flist = os.listdir('./res_gossiping')
```

```python
sb = ''
for f in flist:
    if f == '.ipynb_checkpoints':
        continue
    
    with open('./res_gossiping/' + f, 'r', encoding='utf-8') as file:
        tmp_article_content = file.read()
    
    sb += tmp_article_content.split('---split---')[0]

sb
```

```python
cut_word_list = jieba.cut(sb)
```

```python
for i in cut_word_list:
    print(i)
```

```python
cut_word_list = jieba.cut(sb)
cut_word_list = [i for i in cut_word_list]
cut_word_list
```

```python
word_count = {}

for w in cut_word_list:
    if w in word_count:
        word_count[w] += 1
    else:
        word_count[w] = 1
        
word_count
```

```python
wc = [(k, word_count[k]) for k in word_count]
```

```python
wc.sort(key=lambda t: t[1], reverse=True)
```

```python
jieba.load_userdict('./mydict.txt')

cut_word_list = jieba.cut(sb)
cut_word_list = [i for i in cut_word_list]

word_count = {}

for w in cut_word_list:
    if len(w) <= 1:
        continue
    if w in word_count:
        word_count[w] += 1
    else:
        word_count[w] = 1
        
wc = [(k, word_count[k]) for k in word_count]
wc.sort(key=lambda t: t[1], reverse=True)

wc
```
