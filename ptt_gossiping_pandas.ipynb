```python
import pandas as pd
import os
```

```python
flist = os.listdir('./res_gossiping')
```

```python
columns = ['推', '噓', '分數', '作者', '標題', '時間']
data = []
```

```python
# 製作 data
for f in flist:
    if f == '.ipynb_checkpoints':
        continue
        
    with open('./res_gossiping/' + f, 'r', encoding='utf-8') as file:
        tmp_article_content = file.read()
    
    try:
        tmp_data_str = tmp_article_content.split('---split---')[1]
    except IndexError:
        continue
        
    tmp_data = tmp_data_str.split('\n')[1:-1]
    data.append(tmp_data)
    
#     print(tmp_data)
print(data)
```

```python
df = pd.DataFrame(data=data, columns=columns)
```

```python
df
```

```python
def column_filter(s):
    output = s.split(': ')[1]
    return output
```

```python
for c in columns:
    df[c] = df[c].apply(column_filter)
df
```
